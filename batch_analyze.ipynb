```python
import json

with open('batch_vs_sp.json', 'r') as f:
    content = json.load(f)
```

```python
import numpy as np
import matplotlib.pyplot as plt

bs = np.array(content['batch_size'])
base_throughout = np.array(content['throughout'])
eagle_throughout = np.array(content['eagle_throughout'])
```

```python
fig, ax1 = plt.subplots()

ax1.plot(bs, base_throughout, marker='.', label='base throughout')
ax1.plot(bs, eagle_throughout, marker='.', label='eagle throughout')
ax1.set_xlabel('batch size')
ax1.set_ylabel('thoughout')
ax1.legend()

ax2 = ax1.twinx()
ax2.plot(bs, eagle_throughout/base_throughout, marker='.',color='brown',label='acceleration rate')
ax2.set_ylabel('speed up')
ax2.legend()

plt.title('throughout versus batch_size')
plt.show()
```
